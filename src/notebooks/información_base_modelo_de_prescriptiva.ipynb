```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

R,C=20,20
i,j=np.meshgrid(np.arange(R),np.arange(C))
N=6 #Número de estaciones posibles

def pr(i,j,t,n): #Distribución de probabilidad de personas
  rate=np.sin(2*np.pi*i/20)*np.sin(2*np.pi*j/15)*2+2 #Tasa de crecimiento de población por año
  media=rate*t
  #Voy a suponer una distribución de poisson
  prob = poisson.pmf(n, mu=media)
  return prob

def gen(i,j,t): #Distribución de probabilidad de personas
  rate=np.sin(2*np.pi*i/20)*np.sin(2*np.pi*j/15)*2+4 #Tasa de crecimiento de población por año
  media=rate*t
  #Voy a suponer una distribución de poisson
  n = np.random.poisson(lam=media)
  return n

plt.plot(pr(10,10,30,np.arange(0,100)), label='c1')
plt.plot(pr(5,10,20,np.arange(0,100)), label='c2')
plt.plot(pr(4,20,20,np.arange(0,100)), label='c3')
plt.legend()
```

```python
#Análisis de la población a 20 años por ejemplo
num=gen(i,j,20)
plt.contourf(i,j,num, 50)
plt.colorbar()
plt.scatter(i,j,marker='.', color='r', s=8)
```
